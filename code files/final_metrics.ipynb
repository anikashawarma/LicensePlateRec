```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("performance_log.csv")

plt.plot(df["frame"], df["fps"])
plt.title("FPS over Time")
plt.xlabel("Frame")
plt.ylabel("FPS")
plt.grid()
plt.show()
```

```python
plt.plot(df["det_ms"], label="Detection")
plt.plot(df["ocr_ms"], label="OCR")
plt.plot(df["total_ms"], label="End-to-End")

plt.title("Latency Breakdown")
plt.xlabel("Frame")
plt.ylabel("Latency (ms)")
plt.legend()
plt.grid()
plt.show()
```

```python
avg_vals = df[["det_ms","ocr_ms","total_ms"]].mean()

avg_vals.plot(kind="bar")
plt.title("Average Latency (ms)")
plt.ylabel("Milliseconds")
plt.grid()
plt.show()
```
